```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme()
sns.set_context("notebook", font_scale=1.3, rc={"lines.linewidth": 2.2})
from google.colab import files
```

# **Convergência**

```python
conv10 = pd.read_csv("https://raw.githubusercontent.com/andrefcmoraes/MetodosNumericosAplicados/main/10n.csv")
conv25 = pd.read_csv("https://raw.githubusercontent.com/andrefcmoraes/MetodosNumericosAplicados/main/25n.csv")
conv50 = pd.read_csv("https://raw.githubusercontent.com/andrefcmoraes/MetodosNumericosAplicados/main/50n.csv")
conv100 = pd.read_csv("https://raw.githubusercontent.com/andrefcmoraes/MetodosNumericosAplicados/main/100n.csv")
conv250 = pd.read_csv("https://raw.githubusercontent.com/andrefcmoraes/MetodosNumericosAplicados/main/250n.csv")
conv500 = pd.read_csv("https://raw.githubusercontent.com/andrefcmoraes/MetodosNumericosAplicados/main/500n.csv")
conv1000 = pd.read_csv("https://raw.githubusercontent.com/andrefcmoraes/MetodosNumericosAplicados/main/1000n.csv")

con10 = conv10.to_numpy()
con25 = conv25.to_numpy()
con50 = conv50.to_numpy()
con100 = conv100.to_numpy()
con250 = conv250.to_numpy()
con500 = conv500.to_numpy()
con1000 = conv1000.to_numpy()
```

```python
plt.plot(con10[:,0], con10[:,1], label="10 nós")
plt.plot(con25[:,0], con25[:,1], label="25 nós")
plt.plot(con50[:,0], con50[:,1], label="50 nós")
plt.plot(con100[:,0], con100[:,1], label="100 nós")
plt.plot(con250[:,0], con250[:,1], label="250 nós")
plt.plot(con500[:,0], con500[:,1], label="500 nós")
plt.plot(con1000[:,0], con1000[:,1], label="1000 nós")
plt.ylabel('Temperatura  (°C)')
plt.xlabel('Posição (m)')
plt.legend(loc="upper right")
plt.show()
```
